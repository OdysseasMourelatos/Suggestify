```python
import pandas as pd

# Υπολογισμός ωρών ακρόασης ανά καλλιτέχνη από τη βάση σου
query_top_artists = """
    SELECT 
        a.name AS artist_name, 
        ROUND(SUM(s.ms_played) / 3600000.0, 2) AS hours_played
    FROM streams s
    JOIN song_artists sa ON s.song_id = sa.song_id
    JOIN artists a ON sa.artist_id = a.id
    WHERE sa.is_feature = FALSE
    GROUP BY a.name
    ORDER BY hours_played DESC
    LIMIT 10;
"""

df_top_artists = pd.read_sql(query_top_artists, engine)
df_top_artists
```
